# Ciphers

## Introduction

For centuries, people have tried to encrypt texts so that outsiders would not be able to easily understand them. In this task, you will have to decipher a text created by a group of preschool children who decided to encrypt part of the contents of their parents’ library by covering letters in selected lines with stickers.

## Task

The children started by preparing $n$ bags: one for each unique character in the entire library. The boys drew $3n$ different pictures, and the girls printed stickers with these pictures: each picture was printed in some unknown and possibly different number of copies. However, there are enough stickers that they will not run out during play. Then, into each of the prepared bags, the children put stickers corresponding to exactly three pictures. Altogether, all stickers ended up in one of the bags.

During the game, the children decided to encrypt some lines in the books. For each encountered character, they would take a random sticker from the bag corresponding to that character and place it in its position. As children do, they got bored, so not all lines in the books were processed this way. Unfortunately, the pictures cannot be removed in any way, but perhaps by analyzing the arrangement of letters in the uncovered texts, we will be able to reconstruct the text from the covered lines.

Write a function `decipher_corpus(clear_corpus, ciphered_corpus)` which takes two arguments:

## Evaluation

For this task, you can receive a score between 0 and 1 point. Your function `decipher_corpus(clear_corpus, ciphered_corpus)` will be tested on 4 test cases: independent ciphers. The evaluation criterion for a given cipher will be the percentage of correctly deciphered characters in the text (the number of correctly deciphered characters divided by the total number of characters in the text). A model achieving classification accuracy below 50% will receive 0 points.

The final score will be the average of the results for all tests and will be calculated according to the formula:

$$\mathrm{score} =  \frac{1}{4} \sum_{i=1}^4 2 \cdot \max(\mathrm{accuracy}_i - 0.5, 0)$$

`````python
import numpy as np
from collections import Counter
from k_means_constrained import KMeansConstrained

def decipher_corpus(clear_corpus, ciphered_corpus):

    flat_ciphers = [list(s) for s in ciphered_corpus]

    model = Letter2Vec(flat_ciphers, vector_size=50, min_count=1)

    unique_tokens = sorted({
        t
        for s in flat_ciphers
        for t in s
    })

    embeddings = np.array([
        model.wv[t]
        for t in unique_tokens
    ])

    kmeans = KMeansConstrained(
        n_clusters=len(unique_tokens) // 3,
        size_min=3,
        size_max=3,
        random_state=42
    )

    labels = kmeans.fit_predict(embeddings)

    token_to_cluster = {
        t: c
        for t, c in zip(unique_tokens, labels)
    }

    cipher_clusters = [
        token_to_cluster[t]
        for s in flat_ciphers
        for t in s
    ]

    cipher_counts = Counter(cipher_clusters)

    flat_clear = [
        t
        for s in clear_corpus
        for t in list(s)
    ]

    clear_counts = Counter(flat_clear)

    cluster_sorted = [k for k, _ in cipher_counts.most_common()]
    clear_sorted = [k for k, _ in clear_counts.most_common()]

    n = min(len(cluster_sorted), len(clear_sorted))

    cluster_to_clear = {
        cluster_sorted[i]: clear_sorted[i]
        for i in range(n)
    }

    for i in range(n, len(cluster_sorted)):
        cluster_to_clear[cluster_sorted[i]] = clear_sorted[-1]

    manual_swaps = {
        'r': 's',
        's': 'r',
        ',': 'l',
        'l': ',',
        '”': ':',
        ':': '”'
    }

    cluster_to_clear = {
        k: manual_swaps.get(v, v)
        for k, v in cluster_to_clear.items()
    }

    decoded_corpus = []

    for sentence in ciphered_corpus:
        decoded_sentence = [
            cluster_to_clear[token_to_cluster[emoji]]
            for emoji in sentence
        ]
        decoded_corpus.append(decoded_sentence)

    decoded_sentences = [
        "".join(sentence)
        for sentence in decoded_corpus
    ]

    return decoded_sentences


decoded_corpus = decipher_corpus(corpus_clear, corpus_ciphered)
`````
